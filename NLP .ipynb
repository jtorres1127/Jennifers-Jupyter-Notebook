```python
def text_process(mess): 
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
   
        






```

```python
messages[['Notes / Reason for Non-Compliance']] = messages[['Notes / Reason for Non-Compliance']].astype('str')
```

```python
compliant = messages[(messages['Current Status'] == 'Open') & (messages['Compliance Status'] == 'Yes')].reset_index(drop = True)
```

```python
#messages.arr[0:]


compliant['Notes / Reason for Non-Compliance'].head(100).apply(text_process)
        

#print('Number of rows in Dataframe :', len(messages.index))

#pandas.options.display.max_rows
#print(messages)

#print ('Number of rows in Dataframe :' len(messages.index))





```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
bow_transformer = CountVectorizer(analyzer=text_process).fit(compliant['Notes / Reason for Non-Compliance'])
```
